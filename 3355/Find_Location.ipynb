```python
import pandas as pd
from geopy.geocoders import Nominatim

# CSV 파일 불러오기
csv = pd.read_csv('Dongdaemun_all.csv', encoding='utf-8')

# 위도, 경도 반환하는 함수
def geocoding_by_place_name(place_name):
    try:
        geo = geo_local.geocode(place_name)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0, 0]

# 위도, 경도 반환하는 객체 생성
geo_local = Nominatim(user_agent='South Korea')

# 관광지명을 이용하여 위, 경도 값으로 변환하기
latitude = []
longitude = []

for place_name in csv['관광지명']:
    coordinates = geocoding_by_place_name(place_name)
    latitude.append(coordinates[0])
    longitude.append(coordinates[1])

# DataFrame에 새로운 열 추가
csv['관광지 위도'] = latitude
csv['관광지 경도'] = longitude
```

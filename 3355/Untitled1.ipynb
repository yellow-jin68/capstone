```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=chrome_options)

url = 'https://www.google.com/maps/search/%EC%9D%B8%EC%82%AC%EB%8F%99+%EB%A7%9B%EC%A7%91/data=!3m1!4b1?hl=ko&entry=ttu'
driver.get(url)

# 페이지가 완전히 로드될 때까지 대기 (타임아웃은 10초로 설정)
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
```

```python
from selenium import webdriver

# Chrome WebDriver 초기화
driver = webdriver.Chrome()

# 웹 페이지로 이동
driver.get("https://www.google.com/maps/search/%EC%9D%B8%EC%82%AC%EB%8F%99+%EB%A7%9B%EC%A7%91/data=!3m1!4b1?hl=ko&entry=ttu")

# 현재 웹 페이지의 HTML 소스 코드 가져오기
page_content = driver.page_source

# 페이지 소스를 출력
print(page_content)

# WebDriver 종료
driver.quit()
```
